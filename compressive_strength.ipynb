# Machine Learning Modeling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.ensemble import AdaBoostRegressor
from xgboost import XGBRegressor
from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
```

```python
# Features
X = df.drop('Concrete_compressive_strength', axis='columns')
# Target variable
y = df['Concrete_compressive_strength']
```

```python
X.shape, y.shape
```

```python
# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
X_train.shape, X_test.shape
```

```python
# # scaling the data
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Using Cross Validation

```python
models = {"SVM":SVR(kernel="rbf", degree=2, C=100, epsilon=0.01),
         "DTReg":DecisionTreeRegressor(),
         "RFReg":RandomForestRegressor(),
         "GBR":GradientBoostingRegressor(),
         "HGBR":HistGradientBoostingRegressor(),
         "AdBR":AdaBoostRegressor(),
         "XGBR":XGBRegressor(),
         "CatBoostReg":CatBoostRegressor(verbose=None),
         "LGBM":LGBMRegressor()}

accuracy = []
for name, model in models.items():
    print("Training {}".format(name))
    scores = cross_val_score(model, X_train_scaled, y_train, cv=5)
    print("cross val scores of {} : {}".format(name, scores))
    print("Average score of {} : {}".format(name, scores.mean()))
    print("-----------------------------------------------------")
```

## Support Vector Regressor

```python
svr= SVR(kernel="rbf", degree=2, C=100, epsilon=0.01)
svr.fit(X_train_scaled, y_train)
svr_preds = svr.predict(X_test_scaled)
```

```python
plt.rcParams['figure.figsize'] = (10, 8)
sns.kdeplot(y_test)
sns.kdeplot(svr_preds)
plt.legend(['actual', 'predicted'])
```

```python
print("Training Accuracy: {}".format(svr.score(X_train_scaled, y_train)))
print("Testing Accuracy: {}".format(svr.score(X_test_scaled, y_test)))
print("-------------------------------")
print("R Squared: {}".format(r2_score(y_test, svr_preds)))
print("Root Mean Squared Error: {}".format(np.sqrt(mean_squared_error(y_test, svr_preds))))
print("Mean Absolute Error: {}".format(mean_absolute_error(y_test, svr_preds)))
```

### Evaluation Metrics Table

```python
## Error table
## hgbr, svr, dtc, rf, gbr, cbr, adb, lgbm
## r2-score, mean-squared-error, mean-absolute-error
models = [svr, dt_reg, rf_reg, gbr, hgbr, xgb_reg, cbr, adb, lgbm_reg]
r_squared = []
rmse = []
mae = []
for model in models:
    preds = model.predict(X_test_scaled)
    r_squared.append(r2_score(y_test, preds))
    rmse.append(np.sqrt(mean_squared_error(y_test, preds)))
    mae.append(mean_absolute_error(y_test, preds))
    
error_table = pd.DataFrame(columns=["Models","R_Squared", "RMSE", "MAE"], index= range(len(models)))
algorithms = ["Support Vector Regressor", 
              "DecisionTreeRegressor",
              "RandomForestRegressor", 
              "GradientBoostingRegressor",
              "HistGradientBoostingRegressor",
              "XGBRegressor",
              "CatBoostRegressor",
              "AdaBoostRegressor",
              "LGBMRegressor"]

error_table["Models"] = algorithms
error_table["R_Squared"] = r_squared
error_table["RMSE"] = rmse
error_table["MAE"] = mae   

error_table
    
```
